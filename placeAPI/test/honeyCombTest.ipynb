```python
import matplotlib.pyplot as plt
from math import sqrt

def honeycombSearch(topLeftLat, topLeftLon, downRightLat, downRightLon, radius):

    fig, ax = plt.subplots()
    fig = plt.figure(figsize=(20, 20))
    ax.set_aspect( 1 )

    x = sqrt(3)
    print(x)
    """
    radius = 0.1
    topLeftLat, topLeftLon, downRightLat, downRightLon = 0, 0, 1, 1
    """
    shift = 0

    i = topLeftLat
    while i < downRightLat: 
        
        j = topLeftLon - sqrt(3) * radius * 0.5 *(shift%2)
        while j < downRightLon:
            circle = plt.Circle(( i, j), radius, fill = False)
            ax.add_patch(circle)
            j += sqrt(3) * radius
        shift = (shift+1) % 2
        
        i += 1.5 * radius

honeycombSearch(0,0,1,1,0.1)
```
