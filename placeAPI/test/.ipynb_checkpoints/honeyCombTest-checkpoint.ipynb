```python
import matplotlib.pyplot as plt
from math import sqrt

fig, ax = plt.subplots()
fig = plt.figure(figsize=(20, 20))
ax.set_aspect( 1 )

x = sqrt(3)
print(x)

radius = 0.05

flag = 0
i = 0
while i < 1: 
    j = 0 - sqrt(3) * radius * 0.5 *(flag%2)
    while j < 1:
        circle = plt.Circle(( i, j), radius, fill = False)
        ax.add_patch(circle)
        j += sqrt(3) * radius
    flag = (flag+1) % 2
    
    i += 1.5 * radius
```
